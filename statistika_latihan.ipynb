# 📊 Latihan Statistika Dasar

Notebook ini berisi latihan dasar statistika menggunakan Python:
- Mean, Median, Mode
- Range, Variance, Standard Deviation
- Distribusi Normal & Visualisasi Histogram

Tools: NumPy, Pandas, Seaborn, Matplotlib

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

## 🔢 Dataset Contoh (Tinggi Badan dalam cm)

```python
data = [160, 165, 162, 170, 168, 155, 172, 175, 163, 160]
df = pd.DataFrame(data, columns=['tinggi'])
df
```

## 📈 Ukuran Pusat & Sebaran

```python
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True).mode[0]
range_ = np.max(data) - np.min(data)
variance = np.var(data)
std_dev = np.std(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Range: {range_}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

## 🔍 Visualisasi Distribusi Data

```python
sns.histplot(data, kde=True)
plt.title("Distribusi Tinggi Badan")
plt.xlabel("Tinggi (cm)")
plt.ylabel("Frekuensi")
plt.grid()
plt.show()
```
